##Theoretical Questions

**1. What is unsupervised learning in the context of machine learning?**

Ans-  Unsupervised learing is a type of machine learing Wher algorithm learn from **unlabeled data** without human supervision. The goal is to discover inherent patterns, structures, and relatuionship within the data, such as grouping similar data points (clustering) or reducing the number of features .

**2.How does K-Means clustering algorithm work?**

Ans-   The **K-Mean** algorithm is an iterative ,centroid-based method that partitions a dataset into a pre-specified number of k cluster. It works as follows:

1. Initialization: k cluster centriods are intially selected ,usuall at random.

2. Assigment step: Each data point is assigned to the cluster .

The algorithm starts by randomly choosing K centroids. Each data point is assigned to the nearest centroid. Then, new centroids are calculated as the mean of points in each cluster. This process repeats until centroids stop changing.

**3. Explain the concept of a dendrogram in hierarchical clustering.**

Answer:
A dendrogram is a tree-like diagram used to visualize hierarchical clustering.

Explanation:
It shows how individual data points are merged into clusters step by step. The height of the branches represents the distance at which clusters are joined.

**4. What is the main difference between K-Means and Hierarchical Clustering?**

Answer:
K-Means is a partitioning method, while hierarchical clustering builds a tree structure.

Explanation:
K-Means requires the number of clusters in advance and works iteratively. Hierarchical clustering does not require K initially and shows relationships between clusters but is computationally expensive.

**5. What are the advantages of DBSCAN over K-Means?**

Answer:
DBSCAN can find arbitrary-shaped clusters and detect noise.

Explanation:
Unlike K-Means, DBSCAN does not assume spherical clusters and does not require the number of clusters beforehand.

**6. When would you use Silhouette Score in clustering?**

Answer:
Silhouette Score is used to evaluate clustering quality.

Explanation:
It measures how similar a data point is to its own cluster compared to other clusters, helping choose the optimal number of clusters.

**7. What are the limitations of Hierarchical Clustering?**

Answer:
It is slow and memory-intensive.

Explanation:
Hierarchical clustering does not scale well for large datasets and once clusters are merged or split, the process cannot be reversed.

**8. Why is feature scaling important in clustering algorithms like K-Means?**

Answer:
Feature scaling ensures equal contribution of all features.

Explanation:
Since K-Means relies on distance calculations, features with larger numerical ranges can dominate clustering results if not scaled.

**9. How does DBSCAN identify noise points?**

Answer:
Points with insufficient neighbors are labeled as noise.

Explanation:
If a point does not have at least MinPts within radius ε, it is considered an outlier.

**10. Define inertia in the context of K-Means.**

Answer:
Inertia is the sum of squared distances between points and their centroids.

Explanation:
It measures how compact the clusters are. Lower inertia means better clustering.

**11. What is the elbow method in K-Means clustering?**

Answer:
A method to find the optimal number of clusters.

Explanation:
Inertia is plotted against K. The point where the curve bends (elbow) suggests the best K.

**12. Describe the concept of "density" in DBSCAN.**

Answer:
Density refers to the concentration of points in a region.

Explanation:
Clusters are formed in regions where points are closely packed together based on ε and MinPts.

**13. Can hierarchical clustering be used on categorical data?**

Answer:
Yes, with suitable distance measures.

Explanation:
Distance metrics like Jaccard or Hamming distance allow hierarchical clustering on categorical data.

**14. What does a negative Silhouette Score indicate?**

Answer:
Poor clustering.

Explanation:
It means a data point is closer to another cluster than its own, indicating incorrect cluster assignment.

**15. Explain the term "linkage criteria" in hierarchical clustering.**

Answer:
Linkage criteria define how distance between clusters is calculated.

Explanation:
Examples include single, complete, average, and Ward linkage, each affecting cluster shape.

**16. Why might K-Means clustering perform poorly on data with varying cluster sizes or densities?**

Answer:
Because K-Means assumes equal-sized, spherical clusters.

Explanation:
It cannot properly handle clusters with different shapes, sizes, or densities.

**17. What are the core parameters in DBSCAN, and how do they influence clustering?**

Answer:
ε (epsilon) and MinPts.

Explanation:
ε defines neighborhood size, while MinPts defines minimum points required to form a cluster. They control cluster formation and noise detection.

**18. How does K-Means++ improve upon standard K-Means initialization?**

Answer:
It selects better initial centroids.

Explanation:
Centroids are chosen far apart, reducing chances of poor clustering and improving convergence.

**19. What is agglomerative clustering?**

Answer:
A bottom-up hierarchical clustering approach.

Explanation:
Each data point starts as its own cluster, and clusters are merged step by step based on similarity.

**20. What makes Silhouette Score a better metric than just inertia for model evaluation?**

Answer:
It considers both cohesion and separation.

Explanation:
Inertia measures only compactness, while Silhouette Score also checks how well clusters are separated.

21.Generate synthetic data with 4 centers using make_blobs and apply K-Means clustering. Visualize using a
scatter plot.

```python
# Python code for generating synthetic data with 4 centers, applying K-Means, and visualizing

import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans

# Step 1: Generate synthetic data with 4 centers
X, y_true = make_blobs(n_samples=500, centers=4, cluster_std=0.60, random_state=42)

# Step 2: Apply K-Means clustering
kmeans = KMeans(n_clusters=4, random_state=42)
y_kmeans = kmeans.fit_predict(X)

# Step 3: Visualize the clusters
plt.figure(figsize=(8,6))

# Plot each cluster with unique color
plt.scatter(X[:, 0], X[:, 1], c=y_kmeans, s=50, cmap='viridis', alpha=0.6)

# Plot the cluster centroids
centers = kmeans.cluster_centers_
plt.scatter(centers[:, 0], centers[:, 1], c='red', s=200, alpha=0.9, marker='X')
plt.title("K-Means Clustering of Synthetic Data")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.show()

# Optional: Print cluster centers
print("Cluster centers:", centers)
```

23. Generate synthetic data using make_moons and apply DBSCAN. Highlight outliers in the plot.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_moons
from sklearn.cluster import DBSCAN

# Step 1: Generate synthetic two-moon dataset
X, _ = make_moons(n_samples=300, noise=0.07, random_state=42)

# Step 2: Apply DBSCAN clustering
# eps: neighborhood radius, min_samples: minimum points to form a dense region
dbscan = DBSCAN(eps=0.2, min_samples=5)
labels = dbscan.fit_predict(X)

# Step 3: Identify core points, border points, and outliers
# Outliers are labeled as -1 by DBSCAN
outliers = labels == -1
clusters = labels != -1

# Step 4: Plot results
plt.figure(figsize=(8, 6))

# Plot clusters with different colors
unique_labels = set(labels)
colors = plt.cm.tab10(np.linspace(0, 1, len(unique_labels)))

for label, color in zip(unique_labels, colors):
    if label == -1:
        # Outliers in red
        plt.scatter(X[outliers, 0], X[outliers, 1], c='red', marker='x', s=80, label='Outliers')
    else:
        plt.scatter(X[labels == label, 0], X[labels == label, 1], c=[color], label=f'Cluster {label}')

plt.title("DBSCAN Clustering on make_moons Data")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.legend()
plt.grid(True)
plt.show()
```

25. Use make_circles to generate synthetic data and cluster it using DBSCAN. Plot the result.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_circles
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler

# Step 1: Generate synthetic circular data
X, y_true = make_circles(n_samples=500, factor=0.5, noise=0.05, random_state=42)

# Step 2: Standardize features (important for DBSCAN performance)
X_scaled = StandardScaler().fit_transform(X)

# Step 3: Apply DBSCAN clustering
# eps: neighborhood radius, min_samples: minimum points to form a cluster
dbscan = DBSCAN(eps=0.2, min_samples=5)
labels = dbscan.fit_predict(X_scaled)

# Step 4: Identify core samples and outliers
core_samples_mask = np.zeros_like(labels, dtype=bool)
core_samples_mask[dbscan.core_sample_indices_] = True

# Step 5: Plot the results
unique_labels = set(labels)
colors = plt.cm.Spectral(np.linspace(0, 1, len(unique_labels)))

plt.figure(figsize=(8, 6))
for k, col in zip(unique_labels, colors):
    if k == -1:
        # Outliers are plotted in black
        col = 'k'

    class_member_mask = (labels == k)

    # Plot core points
    xy = X_scaled[class_member_mask & core_samples_mask]
    plt.plot(xy[:, 0], xy[:, 1], 'o', markerfacecolor=col,
             markeredgecolor='k', markersize=8)

    # Plot border points
    xy = X_scaled[class_member_mask & ~core_samples_mask]
    plt.plot(xy[:, 0], xy[:, 1], 'o', markerfacecolor=col,
             markeredgecolor='k', markersize=4)

plt.title("DBSCAN Clustering on make_circles Data")
plt.xlabel("Feature 1 (scaled)")
plt.ylabel("Feature 2 (scaled)")
plt.show()
```

29. Create synthetic data using make_blobs and evaluate silhouette scores for k = 2 to 5. Display as a bar chart.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

# Step 1: Generate synthetic dataset
# 500 samples, 4 centers, 2 features
X, y_true = make_blobs(n_samples=500, centers=4, cluster_std=0.6, random_state=42)

# Step 2: Evaluate silhouette scores for k = 2 to 5
k_values = range(2, 6)
silhouette_scores = []

for k in k_values:
    try:
        kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
        labels = kmeans.fit_predict(X)
        score = silhouette_score(X, labels)
        silhouette_scores.append(score)
    except Exception as e:
        print(f"Error for k={k}: {e}")
        silhouette_scores.append(np.nan)

# Step 3: Plot silhouette scores as a bar chart
plt.figure(figsize=(8, 5))
bars = plt.bar(k_values, silhouette_scores, color='skyblue', edgecolor='black')

# Annotate bars with score values
for bar, score in zip(bars, silhouette_scores):
    plt.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 0.01,
             f"{score:.3f}", ha='center', fontsize=10)

plt.title("Silhouette Scores for Different k Values")
plt.xlabel("Number of Clusters (k)")
plt.ylabel("Silhouette Score")
plt.ylim(0, 1)
plt.xticks(k_values)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

30.  Load the Iris dataset and use hierarchical clustering to group data. Plot a dendrogram with average linkage.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.preprocessing import StandardScaler
from scipy.cluster.hierarchy import linkage, dendrogram

# Step 1: Load the Iris dataset
iris = load_iris()
X = pd.DataFrame(iris.data, columns=iris.feature_names)

# Step 2: Standardize the features (important for clustering)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Step 3: Perform hierarchical clustering with average linkage
# 'average' linkage computes the average distance between clusters
Z = linkage(X_scaled, method='average')

# Step 4: Plot the dendrogram
plt.figure(figsize=(12, 6))
dendrogram(
    Z,
    labels=iris.target,  # Optional: use species labels
    leaf_rotation=90,
    leaf_font_size=10
)
plt.title("Hierarchical Clustering Dendrogram (Average Linkage)")
plt.xlabel("Sample Index or Cluster Label")
plt.ylabel("Distance")
plt.tight_layout()
plt.show()
```

32.  Load the Digits dataset and apply DBSCAN after reducing dimensions with t-SNE. Visualize the results.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_digits
from sklearn.manifold import TSNE
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler

# Step 1: Load the Digits dataset
digits = load_digits()
X = digits.data
y_true = digits.target  # True labels (for reference, not used in clustering)

# Step 2: Standardize features (important for DBSCAN)
X_scaled = StandardScaler().fit_transform(X)

# Step 3: Reduce dimensions with t-SNE
print("Running t-SNE dimensionality reduction...")
tsne = TSNE(n_components=2, random_state=42, perplexity=30, n_iter=1000)
X_tsne = tsne.fit_transform(X_scaled)

# Step 4: Apply DBSCAN clustering
print("Applying DBSCAN clustering...")
dbscan = DBSCAN(eps=3, min_samples=5)  # eps may need tuning
labels = dbscan.fit_predict(X_tsne)

# Step 5: Visualization
plt.figure(figsize=(10, 7))
unique_labels = set(labels)
colors = plt.cm.tab10(np.linspace(0, 1, len(unique_labels)))

for label, color in zip(unique_labels, colors):
    mask = labels == label
    if label == -1:
        # Noise points
        plt.scatter(X_tsne[mask, 0], X_tsne[mask, 1], c='k', marker='x', label='Noise')
    else:
        plt.scatter(X_tsne[mask, 0], X_tsne[mask, 1], c=[color], label=f'Cluster {label}')

plt.title("DBSCAN Clustering on Digits Dataset (t-SNE reduced)")
plt.xlabel("t-SNE Feature 1")
plt.ylabel("t-SNE Feature 2")
plt.legend()
plt.show()
```

41. Generate synthetic blobs with 5 centers and apply KMeans. Then use silhouette_score to evaluate the
clustering.

```python
import numpy as np
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt

# Step 1: Generate synthetic data with 5 centers
n_samples = 1000
n_features = 2
n_clusters = 5
random_state = 42

X, y_true = make_blobs(
    n_samples=n_samples,
    n_features=n_features,
    centers=n_clusters,
    cluster_std=1.0,
    random_state=random_state
)

# Step 2: Apply KMeans clustering
try:
    kmeans = KMeans(
        n_clusters=n_clusters,
        init='k-means++',
        n_init=10,
        max_iter=300,
        random_state=random_state
    )
    labels = kmeans.fit_predict(X)
except Exception as e:
    print(f"Error during KMeans clustering: {e}")
    exit(1)

# Step 3: Evaluate using silhouette score
try:
    score = silhouette_score(X, labels)
    print(f"Silhouette Score for {n_clusters} clusters: {score:.4f}")
except Exception as e:
    print(f"Error calculating silhouette score: {e}")
    exit(1)

# Step 4: Visualize the clusters
plt.figure(figsize=(8, 6))
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis', s=30, alpha=0.7, edgecolors='k')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1],
            c='red', marker='X', s=200, label='Centroids')
plt.title(f"KMeans Clustering with {n_clusters} Centers\nSilhouette Score: {score:.4f}")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.legend()
plt.show()
```

44. Load the Iris dataset and plot the Silhouette Coefficient for each sample after KMeans clustering.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_samples, silhouette_score

# Load the Iris dataset
iris = datasets.load_iris()
X = iris.data
y = iris.target  # Not used for clustering, only for reference

# Number of clusters (K)
n_clusters = 3

# Perform KMeans clustering
kmeans = KMeans(n_clusters=n_clusters, random_state=42, n_init=10)
cluster_labels = kmeans.fit_predict(X)

# Compute the average silhouette score
avg_silhouette = silhouette_score(X, cluster_labels)
print(f"Average Silhouette Score: {avg_silhouette:.3f}")

# Compute silhouette scores for each sample
sample_silhouette_values = silhouette_samples(X, cluster_labels)

# Plot silhouette scores
fig, ax = plt.subplots(figsize=(8, 6))
y_lower = 10  # Starting position for the first cluster's plot

for i in range(n_clusters):
    # Aggregate silhouette scores for samples in cluster i
    ith_cluster_silhouette_values = sample_silhouette_values[cluster_labels == i]
    ith_cluster_silhouette_values.sort()

    size_cluster_i = ith_cluster_silhouette_values.shape[0]
    y_upper = y_lower + size_cluster_i

    color = plt.cm.nipy_spectral(float(i) / n_clusters)
    ax.fill_betweenx(
        np.arange(y_lower, y_upper),
        0,
        ith_cluster_silhouette_values,
        facecolor=color,
        edgecolor=color,
        alpha=0.7
    )

    # Label the silhouette plots with cluster numbers
    ax.text(-0.05, y_lower + 0.5 * size_cluster_i, str(i))
    y_lower = y_upper + 10  # Add space between clusters

# Draw the average silhouette score line
ax.axvline(x=avg_silhouette, color="red", linestyle="--")

# Formatting the plot
ax.set_title("Silhouette Plot for KMeans Clustering on Iris Dataset")
ax.set_xlabel("Silhouette Coefficient Values")
ax.set_ylabel("Cluster Label")
ax.set_yticks([])  # No y-axis labels
ax.set_xlim([-0.1, 1])
ax.set_ylim([0, len(X) + (n_clusters + 1) * 10])

plt.show()
```

47.  Generate noisy blobs using make_blobs and use DBSCAN to identify both clusters and noise points. Print the
count.

```python
import numpy as np
from sklearn.datasets import make_blobs
from sklearn.cluster import DBSCAN
import matplotlib.pyplot as plt

# Step 1: Generate synthetic data with noise
# ------------------------------------------
# n_samples: total points
# centers: number of cluster centers
# cluster_std: standard deviation of clusters
# random_state: reproducibility
X, y_true = make_blobs(
    n_samples=300,
    centers=3,
    cluster_std=0.60,
    random_state=42
)

# Add random noise points
rng = np.random.RandomState(42)
noise_points = rng.uniform(low=-6, high=6, size=(20, 2))
X = np.vstack([X, noise_points])

# Step 2: Apply DBSCAN clustering
# --------------------------------
# eps: neighborhood radius
# min_samples: minimum points to form a dense region
dbscan = DBSCAN(eps=0.8, min_samples=5)
labels = dbscan.fit_predict(X)

# Step 3: Identify clusters and noise
# ------------------------------------
n_clusters = len(set(labels)) - (1 if -1 in labels else 0)
n_noise = list(labels).count(-1)

print(f"Number of clusters found: {n_clusters}")
print(f"Number of noise points: {n_noise}")

# Print cluster assignments for first 20 points
print("\nSample cluster assignments (label -1 means noise):")
for i, label in enumerate(labels[:20]):
    print(f"Point {i}: Cluster {label}")

# Step 4: Visualize results
# --------------------------
plt.figure(figsize=(8, 6))
unique_labels = set(labels)
colors = plt.cm.tab10(np.linspace(0, 1, len(unique_labels)))

for k, col in zip(unique_labels, colors):
    if k == -1:
        # Noise points in black
        col = [0, 0, 0, 1]
    class_member_mask = (labels == k)
    plt.scatter(X[class_member_mask, 0], X[class_member_mask, 1],
                c=[col], edgecolor='k', s=50, label=f"Cluster {k}" if k != -1 else "Noise")

plt.title(f"DBSCAN Clustering (Clusters: {n_clusters}, Noise: {n_noise})")
plt.legend()
plt.show()
```

48.  Load the Digits dataset, reduce dimensions using t-SNE, then apply Agglomerative Clustering and plot the
clusters.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_digits
from sklearn.manifold import TSNE
from sklearn.cluster import AgglomerativeClustering
from sklearn.preprocessing import StandardScaler

# Step 1: Load the Digits dataset
digits = load_digits()
X = digits.data
y = digits.target  # True labels (for reference, not used in clustering)

# Step 2: Standardize features for better t-SNE performance
X_scaled = StandardScaler().fit_transform(X)

# Step 3: Apply t-SNE for dimensionality reduction to 2D
try:
    tsne = TSNE(n_components=2, random_state=42, perplexity=30, n_iter=1000)
    X_tsne = tsne.fit_transform(X_scaled)
except Exception as e:
    print(f"Error during t-SNE: {e}")
    exit(1)

# Step 4: Apply Agglomerative Clustering
try:
    n_clusters = len(np.unique(y))  # 10 clusters for digits 0-9
    agg_clust = AgglomerativeClustering(n_clusters=n_clusters, linkage='ward')
    cluster_labels = agg_clust.fit_predict(X_tsne)
except Exception as e:
    print(f"Error during clustering: {e}")
    exit(1)

# Step 5: Plot the clustering results
plt.figure(figsize=(10, 6))
scatter = plt.scatter(
    X_tsne[:, 0], X_tsne[:, 1],
    c=cluster_labels, cmap='tab10', s=50, alpha=0.7, edgecolors='k'
)
plt.title("Digits Dataset - t-SNE + Agglomerative Clustering", fontsize=14)
plt.xlabel("t-SNE Component 1")
plt.ylabel("t-SNE Component 2")
plt.colorbar(scatter, label="Cluster Label")
plt.grid(True)
plt.show()
```
